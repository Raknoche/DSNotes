# Task one

Integrate e^(-x^2) from 0 to infinity

```python
import numpy as np
import scipy.integrate

f = lambda x: np.exp(-x**2)
scipy.integrate.quad(f,0,np.inf)
```

# Task two

1) Use numpy to create a sparse array of x data from 0 to 10.  

2) Set the Y data to equal the sin(x). 

3) Do a linear interpolation of X data on a finer scale, and plot the original and interpolated data

4) Do a cubic spline interpolation, including extrapolation

```python
x = np.linspace(0,10,30)
y = np.sin(x)

import scipy.interpolate
interp_func=scipy.interpolate.interp1d(x,y,kind='linear')
xx = np.linspace(0,10,200)
yy = interp_func(xx)

from matplotlib import pyplot as plt
%matplotlib inline
plt.scatter(x,y)
plt.scatter(xx,yy,color='r',alpha=0.2)
plt.show()
```

```python
import scipy.interpolate

interp_func = scipy.interpolate.CubicSpline(x,y,extrapolate=True)
yy=interp_func(xx)
plt.scatter(x,y)
plt.scatter(xx,yy,color='r',alpha=0.2)
plt.show()
```

# Task Three

1) Use numpy to create data for f(x,y) = x^2 + y

2) Do a 2d interpolation of f(x,y)

```python
x = y = np.linspace(0,10,50)
X,Y = np.meshgrid(x,y)
Z = np.sin(X**2 + Y**2)

interp_func = scipy.interpolate.interp2d(X,Y,Z,kind='cubic')
```

# Task Five

1) Create data for a general, multivariate curve with 3 parameters

2) Fit the curve

```python
x = y = np.linspace(-5,5,20)
f = lambda x,a,b,c: a*x[0]**2 + b*x[1]*2 + c
z = f((x,y),5.3,10,22.4)

import scipy.optimize
params, cov = scipy.optimize.curve_fit(f,(x,y),z)

params
```
