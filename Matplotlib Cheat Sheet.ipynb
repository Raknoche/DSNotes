# Matplotlib

First, import pyplot from matplotlib and call it `plt`.

We can also set a number of defaults such as figure size and font size

```python
import numpy as np
from matplotlib import pyplot as plt
import matplotlib
%matplotlib inline
matplotlib.rc('figure', figsize = (14, 7))
matplotlib.rc('font', size = 14)
```

We use matplotlib.pyplot to make plots.  To do so, we first make a figure and axis object, using <pre><b>`f,ax = plt.subplots()`</b></pre>

Next, we use the axis object to draw a plot.  There are many types of plots, including:


| Plot          | Syntax        |  Notes |  
| ------------- |:-------------:| ----- |
| Scatter Plot  | ax.scatter()  | |
| Line Plot     | ax.plot()     | |
| Histogram     | ax.hist()     | |
| 2D Histogram  | ax.hist2d()  | Use to make density plots from X,Y vectors <br> Has count, xedge, yedge, image output  |
| Show Image    |  ax.imshow() | Use to make density plots from 2D matrix |
| Step plot     | ax.step()    | |
| Bar           | ax.bar()      | |
| Errorbar      | ax.errorbar() | |
| Box plot      | ax.boxplot()  | |
| Manual Fill   | ax.fill_between() | Use to fill confidence intervals |
| Contour plot | ax.contour() | Can be used to take numpy.histogram2d output to make smoother contours |
| Filled Contours | ax.contourf() |Can be used to take numpy.histogram2d output to make smoother contours |
    
Some useful axis methods include:

* **`ax.set_title(TITLE)`** sets title
    
* **`ax.set_xlabel(LABEL)`** sets xlabel
    
* **`ax.set_ylabel(LABEL)`** sets ylabel

* **`ax.legend([LEGEND])`** sets legend
    
* **`ax.text(X,Y,TEXT)`** adds text at (X,Y)
    
* **`ax.axis([XMIN,XMAX,YMIN,YMAX])`** sets axis range

* **`ax.xaxis.set_ticks([])`** sets x ticks. Works similarly for yticks
    
* **`ax.grid(True)`** turns on grid

* **`ax.semilogy()`**

* **`ax.semilogx()`**

Some useful pyplot functions are included below.  Note that these operate on the current axis and figure:

* **`plt.figure()`** creates a new figure
* **`ax = plt.add_subplot()`** adds subplots to figure
* **`ax = plt.gca()`** gets current axis
* **`plt.title()`** sets title
* **`plt.xlabel()`** sets xlabel
* **`plt.ylabel()`** sets ylabel
* **`plt.legend()`** sets legend
* **`cbar = plt.colorbar()`** adds colorbar.
* **`cbar.ax.set_ylabel(LABEL,rotation=270)`**  labels the colorbar.
* **`plt.xtick()`** sets xticks
* **`plt.ytick()`** sets yticks
* **`plt.autoscale(enable=True, axis='x', tight=True)`** automatically set margins to be small on an axis
* **`cmap = plt.cm.get_cmap(MAP_NAME)`** save a colormap to a variable


A few additional notes:

* You can create equations using raw strings and latex's dollar format

* You can overlay multiple plots on the same figure (essentially, MATLAB's hold is always on)

* **`plt.clf()`** clears the current figure

* You can set the min and max of the colorbar range using the **`vmin`** and **`vmax`** arguments. 




```python
# Example
x_data=np.linspace(0,10,50)
y_data=x_data**2

f,ax = plt.subplots()
ax.scatter(x_data,y_data,s = 30, color = '#539caf', alpha = 0.75)
ax.set_title('My Title')
ax.set_xlabel('My X Label')
ax.set_ylabel('My Y Label')
ax.grid(True)
```

# Density plots

```python
# Generate fake data
x = np.random.normal(size=1000)
y = x * 3 + np.random.normal(size=1000)
```

```python
#Contour plot from hist2d -- SOMETHING IS OFF WITH EXTENT

#histogram2d is the same as hist2d, without plotting
counts, xedges, yedges = histogram2d(x, y, bins=20, range=None, normed=False)
x_bin_center=(xedges[:-1]+xedges[1:])/2
y_bin_center=(yedges[:-1]+yedges[1:])/2

plt.clf()
cmap = plt.cm.get_cmap("BuGn")
cmap.set_under("w")
plt.contour(x_bin_center,y_bin_center,counts.transpose(),colors="g")
plt.contourf(x_bin_center,y_bin_center,counts.transpose(),cmap=cmap,vmin=6)
plt.colorbar()
plt.scatter(x,y,color='k',alpha=0.07)
```

```python
#Scatter plot colored by density
from scipy.stats import gaussian_kde

# Calculate the point density
xy = np.vstack([x,y])
z = gaussian_kde(xy)(xy)

# Sort the points by density, so that the densest points are plotted last
idx = z.argsort()
xx, yy, zz = x[idx], y[idx], z[idx]

fig, ax = plt.subplots()
ax.scatter(xx, yy, c=zz, s=50, edgecolor='')
plt.show()
```

# 3D Plots

```python
#To make 3D plots, we have to import Axes3D from mpl_toolkits
from mpl_toolkits.mplot3d import Axes3D

def fun(x, y):
    return x**2 + y

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x = np.linspace(-3.0, 3.0, 10)
y = np.linspace(-4,4,10)
X, Y = np.meshgrid(x, y)
Z = fun(X,Y)

ax.scatter(X, Y, Z)
plt.show()
```

```python
#We can also plot the full surface, rather than scatter plots
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x = np.linspace(-3.0, 3.0, 10)
y = np.linspace(-4,4,10)
X, Y = np.meshgrid(x, y)
Z = fun(X,Y)

ax.plot_surface(X,Y,Z,rstride=1,cstride=1,cmap="Blues")
plt.show()
```
